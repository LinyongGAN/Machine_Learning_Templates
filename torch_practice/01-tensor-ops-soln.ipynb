```python
import torch
import random
import matplotlib.pyplot as plt
%matplotlib inline

# Fix random seeds
random.seed(42)
torch.manual_seed(42)
print("PyTorch version:", torch.__version__)
```

## 1.1 Tensor Creation

```python
# Solution: create tensor x (1,2,3,4), float32 on CPU and print it
import torch
x = torch.arange(24, dtype=torch.float32, device='cpu').view(1,2,3,4)
print('x:', x)
print('shape:', x.shape, 'dtype:', x.dtype, 'device:', x.device)
```

```python
# Solution: change dtype to long
x = x.to(dtype=torch.long)
print('x dtype changed to:', x.dtype)
print(x)
```

```python
# Solution: transpose to (1,2,4,3) by swapping the last two dims
x = torch.transpose(x, -1, -2)
print('shape after transpose:', x.shape)
print(x)
```

```python
# Solution: permute to (4,3,2,1) in one line
# current shape is (1,2,4,3); the order [2,3,1,0] yields sizes [4,3,2,1]
x = x.permute(2,3,1,0)
print('shape after permute:', x.shape)
print(x)
```

```python
# Solution: view to (12, 2)
x = x.view(12, 2)
print('new size:', x.size())
print(x)
```

## Tensor Indexing

```python
# Tensor Indexing
# - Take the slice at index 1 on axis 0
# - Take all slices on axis 1 (:).
# - Take every other slice starting from index 1 on axis 2 (1::2).
import torch
y = torch.arange(24).view(2,3,4)
sel = y[1, :, 1::2]
print('y shape:', y.shape)
print('selected (y[1, :, 1::2]) shape:', sel.shape)
print(sel)
```

## Tensor Arithmetics

```python
# Solution: broadcasting with repeat/view/expand
x = torch.tensor([[1,2],[3,4]])
y = torch.tensor([10,20])

# repeat and reshape to (2,2)
z = y.repeat(2).view(2,2)
print('x:\n', x)
print('y:', y)
print('z (repeat->view to 2x2):\n', z)
print('x+z:\n', x+z)

# view to (2,1) and (1,2) for broadcasting
y_col = y.view(2,1)
y_row = y.view(1,2)
print('x + y_col (broadcast over columns):\n', x + y_col)
print('x + y_row (broadcast over rows):\n', x + y_row)

# do the above using expand (no data copy)
y_col_exp = y.view(2,1).expand(-1,2)
y_row_exp = y.view(1,2).expand(-1,2)
### Note: you may also use .expand_as(x)
print('x + y_col_exp:\n', x + y_col_exp)
print('x + y_row_exp:\n', x + y_row_exp)
```
